```python
import pandas as pd
from transformers import pipeline

df = pd.read_csv('/content/miami_transit_tweets.csv', names=["Tweet_ID", "Author_ID", "Created_At", "Text"])

sentiment_pipeline = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")

def clean_text(text):
    text = str(text)
    text = text.replace('\n', ' ').replace('\r', ' ')
    return text

df['Cleaned_Text'] = df['Text'].apply(clean_text)

def get_bert_sentiment(text):
    result = sentiment_pipeline(text[:512])[0]
    return result['label'], result['score']

df['BERT_Label'], df['BERT_Score'] = zip(*df['Cleaned_Text'].apply(get_bert_sentiment))

df.to_csv('miami_transit_tweets_with_bert.csv', index=False)

print("Saved as 'miami_transit_tweets_with_bert.csv'.")
print(df[['Cleaned_Text', 'BERT_Label', 'BERT_Score']].head())
```
